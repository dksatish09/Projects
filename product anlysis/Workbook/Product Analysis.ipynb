```python
for i in num_df.columns:
    plt.figure(figsize=(5,5))
    sns.boxplot(data=df,x=i)
    plt.show()
```

we found outliers in the mrp and sellprice

# outlier treatment

```python
for i in ['MRP','SellPrice']:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    lc=q1-1.5*iqr
    uc=q3+1.5*iqr
    df[i]=df[i].clip(lower=lc,upper=uc)
```

# Top Brands Based on Orders and Sales

```python
d2=df.groupby(by=['BrandName']).agg(No_of_orders=('BrandName','count'),Total_sales=('SellPrice',np.sum)).sort_values(by='No_of_orders',
                                                                                                                  ascending=False).reset_index().head(10)

d2
```

```python
 plt.figure(figsize=(10,5))
sns.scatterplot(data=d2,x='No_of_orders',y='Total_sales',hue='BrandName',marker="*",sizes=(50,100),s=350)
plt.title("Top Brands Based on Orders and sales")
plt.legend(loc='lower right',)
plt.savefig("Top Brands by Orders and Sales")
plt.show()
```
